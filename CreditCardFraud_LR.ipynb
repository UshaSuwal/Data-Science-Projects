```python
X=df.drop(["Class"],axis=1)
X
```

```python
Y=df[["Class"]]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 12)
```

```python
# Checking length
total_len = (len(X))
print("All total length = ",total_len)    # total row
print("Train length= ", len(X_train))   #total train dataset length
print("Test length = ", len(X_test))  # total test dataset length
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
###### Accuracy
from sklearn.metrics import accuracy_score
predicted = model.predict(X_test)
print("Accuracy=",accuracy_score(Y_test,predicted))
# print('\n')
```

```python
ytest = np.array(Y_test)
print("Classification Report:\n",classification_report(ytest,model.predict(X_test)))
print("\n\nConfusion matrix:\n",confusion_matrix(model.predict(X_test),ytest))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(model.predict(X_test),ytest)
fig, ax = plt.subplots(figsize=(6,6), dpi=90)
class_names = ['Valid','Fraud']
display = ConfusionMatrixDisplay(cm, display_labels=class_names)
ax.set(title='Confusion Matrix')
display.plot(ax=ax);
```
